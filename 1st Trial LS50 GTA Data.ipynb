```python
#HEATMAPS MY FAVE

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('/Users/kyra/Desktop/G9 GTA - Sheet3.csv', skiprows=4)
df.rename(columns={df.columns[0]: 'Reading', df.columns[1]: 'Time (s)'}, inplace=True)
df['Time (s)'] = pd.to_numeric(df['Time (s)'], errors='coerce')
df = df.dropna(subset=['Time (s)'])
well_ids_cleaned = df.columns[2:].tolist()

#od values to numeric
for well in well_ids_cleaned:
    df[well] = pd.to_numeric(df[well], errors='coerce')

#compute growth metrics
min_od = df[well_ids_cleaned].min()
final_od = df[well_ids_cleaned].iloc[-1]
regrowth_magnitude = final_od - min_od

#reshape into 8 x 12
regrowth_matrix = np.array([regrowth_magnitude[well] for well in well_ids_cleaned]).reshape((8, 12))

#column names
strain_map = ['WT', 'WT', 'ΔrogA', 'ΔrogA', 'ΔGTA', 'ΔGTA',
              'WT', 'WT', 'ΔrogA', 'ΔrogA', 'ΔGTA', 'ΔGTA']

#imma group these heatmaps up
grouped_data = {
    'WT': regrowth_matrix[:, [0, 1, 6, 7]],
    'ΔrogA': regrowth_matrix[:, [2, 3, 8, 9]],
    'ΔGTA': regrowth_matrix[:, [4, 5, 10, 11]]
}

fig, axs = plt.subplots(1, 3, figsize=(18, 6), sharey=True)
for ax, (strain, matrix) in zip(axs, grouped_data.items()):
    sns.heatmap(matrix, ax=ax, annot=True, fmt=".2f", cmap='BuGn',
                cbar_kws={'label': 'Regrowth (Final - Min)'}, 
                xticklabels=['CR30', 'CBK', 'CR30', 'CBK'], yticklabels=list("ABCDEFGH"))
    ax.set_title(f"{strain}")
    ax.set_xlabel("Phage / Replicate")
    ax.set_ylabel("Dilution Row")

plt.suptitle("Grouped Heatmaps of Regrowth by Strain (WT, ΔrogA, ΔGTA)", fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
